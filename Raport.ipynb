## Spadochroniarz

Skoki spadochronowe stanowią fascynującą dziedzinę ekstremalnych aktywności, w której łączą się odwaga, precyzja i fizyka. Podczas gdy dla wielu skoków spadochronowych najważniejsze są emocje i adrenalina, z punktu widzenia matematyki możemy spojrzeć na ten ekstremalny sport z zupełnie innej perspektywy. W tym projekcie skoncentrujemy się na analizie ruchu spadochroniarza z wykorzystaniem równań różniczkowych, aby zrozumieć i opisać matematycznie dynamikę tego niezwykłego doświadczenia.
Równanie różniczkowe opisujące ruch spadochroniarza:
$$
ma \quad = \quad F_{g} \quad – \quad F_{oporu}
$$
Na spadochroniarza działają siła grawitacji i siła oporu powietrza.
Najważniejszym aspektem jest siła oporu powietrza. Występuje, gdy ciało porusza się 
w powietrzu. Siła działa przeciwnie do kierunku ruchu ciała i hamuje jego ruch. Jej wielkość zależy od kilku czynników: 
- powierzchnia ciała - im większa powierzchnia ciała, tym większy opór, ponieważ cząsteczki powietrza mają większą powierzchnię do pokonania, aby oderwać się od poruszającego się ciała;
- położenie ciała względem strug powietrza – ustawienie ciała względem kierunku ruchu;
- kształt ciała - aerodynamiczny kształt ciała ma istotny wpływ na siłę oporu. Ciała o bardziej opływowych kształtach generują mniejszy opór niż te o kształtach bardziej niekorzystnych aerodynamicznie;
- gęstość powietrza - im rzadsze powietrze, tym mniejsza siła oporu;
- prędkość ruchu - siła oporu powietrza rośnie proporcjonalnie do kwadratu prędkości ruchu ciała.

W dalszej części znajdziemy rozwiązanie równania $v(t)$. 
Zastosujemy metodę numeryczną i analityczną. Następnie porównamy obydwie metody.


Najpierw rozwiążemy nasze równanie analitycznie - metodą rozdzielenia zmiennych. Nasze równanie ma postać:
$$
m \frac{dv}{dt} = mg - kv^2
$$
Dzieląc je obustronnie przez m otrzymamy:
$$
\frac{dv}{dt} = g - \frac{k}{m} v^2
$$
Następnie, rozdzielając nasze zmienne i całkując obustronnie, otrzymujemy:
$$
\int{\frac{dv}{g - \frac{k}{m} v^2}} = \int{dt}.
$$
Rozkładając wyrażenie po lewej na ułamki proste, otrzymujemy:
$$
\frac{1}{2 \sqrt{g}} \int{\frac{1}{\sqrt{g} - \sqrt{\frac{k}{m}} v} dv} + \frac{1}{2 \sqrt{g}} \int{\frac{1}{\sqrt{g} + \sqrt{\frac{k}{m}} v} dv} = \int{dt}.
$$
Następnie, stosując podstawienia:
$$
x=\sqrt{g} - \sqrt{\frac{k}{m}} v, \quad
dx = -\sqrt{\frac{k}{m}} dv, \quad
y=\sqrt{g} + \sqrt{\frac{k}{m}} v, \quad
dy = \sqrt{\frac{k}{m}} dv
$$
i całkując obustronnie, otrzymujemy:
$$
-\frac{\sqrt{\frac{m}{k}}}{2 \sqrt{g}} \cdot \text{ln} \left| \sqrt{g} - \sqrt{\frac{k}{m}} v \right| \quad + \quad \frac{\sqrt{\frac{m}{k}}}{2 \sqrt{g}} \cdot \text{ln} \left| \sqrt{g} + \sqrt{\frac{k}{m}} v \right| = t + C.
$$
Można to również zapisać jako:
$$
\frac{\sqrt{\frac{m}{k}}}{2 \sqrt{g}}  \cdot \text{ln} \left| \frac{\sqrt{g}+\sqrt{\frac{k}{m}}v}{\sqrt{g}-\sqrt{\frac{k}{m}}v} \right| = t + C,
$$
a po przekształceniu:
$$
\text{ln} \left| \frac{\sqrt{g}+\sqrt{\frac{k}{m}}v}{\sqrt{g}-\sqrt{\frac{k}{m}}v} \right| = 2 \sqrt{\frac{kg}{m}} t + D.
$$
Przekształcając wyrażenie dalej i przy założeniu, że $\sqrt{g}-\sqrt{\frac{k}{m}}v$ >0 ($\sqrt{g}+\sqrt{\frac{k}{m}}v$ również jest większe od 0), otrzymujemy:
$$
\frac{\sqrt{g}+\sqrt{\frac{k}{m}}v}{\sqrt{g}-\sqrt{\frac{k}{m}}v} = \text{e}^{2 \sqrt{\frac{kg}{m}} t + D}.
$$
W ostatnim kroku wyznaczymy prędkość $v$:

$$
\sqrt{g}+\sqrt{\frac{k}{m}}v = \left( \text{e}^{2 \sqrt{\frac{kg}{m}} t + D} \right) \left( \sqrt{g}-\sqrt{\frac{k}{m}}v \right)
$$

$$
\sqrt{g}+\sqrt{\frac{k}{m}}v = \sqrt{g} \cdot \left( \text{e}^{2 \sqrt{\frac{kg}{m}} t + D} \right) - \sqrt{\frac{k}{m}}v \cdot \left( \text{e}^{2 \sqrt{\frac{kg}{m}} t + D} \right)
$$

$$
\sqrt{\frac{k}{m}}v \cdot \left( \text{e}^{2 \sqrt{\frac{kg}{m}} t + D} \right) + \sqrt{\frac{k}{m}}v = \sqrt{g} \cdot \left( \text{e}^{2 \sqrt{\frac{kg}{m}} t + D} \right) - \sqrt{g}
$$
Ostatecznie:
$$
v(t) = \frac{\sqrt{g} \cdot \left( \text{e}^{2 \sqrt{\frac{kg}{m}} t + D} - 1 \right)}{\sqrt{\frac{k}{m}} \cdot \left( \text{e}^{2 \sqrt{\frac{kg}{m}} t + D} + 1 \right)} = \sqrt{\frac{mg}{k}} \cdot \frac{\text{e}^{2 \sqrt{\frac{kg}{m}} t} \cdot D - 1}{\text{e}^{2 \sqrt{\frac{kg}{m}} t} \cdot D + 1},
$$
gdzie $D$ oznacza stałą całkowania, zależną od warunku początkowego. W naszym projekcie będziemy przyjmować, że v(0) = 0. Zatem:
$$
v(0) = \sqrt{\frac{mg}{k}} \cdot \frac{\text{e}^{2 \sqrt{\frac{kg}{m}} \cdot 0} \cdot D - 1}{\text{e}^{2 \sqrt{\frac{kg}{m}} \cdot 0} \cdot D + 1} = \sqrt{\frac{mg}{k}} \cdot \frac{1 \cdot D - 1}{1 \cdot D + 1} = \sqrt{\frac{mg}{k}} \cdot \frac{D - 1}{D + 1} = 0.
$$
Z tego wynika, że D = 1, zatem równanie opisujące ruch skoczka przyjmuje postać:
$$
v(t) = \sqrt{\frac{mg}{k}} \cdot \frac{\text{e}^{2 \sqrt{\frac{kg}{m}} t} - 1}{\text{e}^{2 \sqrt{\frac{kg}{m}} t} + 1}.
$$

Wprowadzimy teraz pojęcie prędkości granicznej. Na spadochroniarza działają dwie siły: grawitacji i oporu. Wartość siły grawitacji nie zmienia się w czasie, natomiast wartość siły oporu rośnie wraz ze wzrostem prędkości spadochroniarza. Kiedy siła oporu osiąga wartość równą wartości siły grawitacji, ich wypadkowa jest równa 0. Oznacza to, że ciało porusza się ruchem jednostajnym, a jego prędkość to prędkość graniczna [1]. 
Aby obliczyć prędkość graniczną, przyjmiemy, że czas dąży do nieskończoności:
$$
v_{gr} = \lim_{t\to\infty} v(t) = \lim_{t\to\infty} \sqrt{\frac{mg}{k}} \cdot \frac{\text{e}^{2 \sqrt{\frac{kg}{m}} t} - 1}{\text{e}^{2 \sqrt{\frac{kg}{m}} t} + 1}.
$$
Aby obliczyć granicę tego wyrażenia, przekształcimy je:
$$
v_{gr} = \lim_{t\to\infty} \sqrt{\frac{mg}{k}} \cdot \frac{\text{e}^{2 \sqrt{\frac{kg}{m}} t} + 1 - 2}{\text{e}^{2 \sqrt{\frac{kg}{m}} t} + 1} = \lim_{t\to\infty} \sqrt{\frac{mg}{k}} \cdot \left( 1 - \frac{2}{\text{e}^{2 \sqrt{\frac{kg}{m}} t} + 1} \right).
$$
Wiemy, że:
$$
\lim_{t\to\infty} \frac{2}{\text{e}^{2 \sqrt{\frac{kg}{m}} t} + 1} = 0, 
$$
ponieważ:
$$
\lim_{t\to\infty} \text{e}^{2 \sqrt{\frac{kg}{m}} t} = \infty.
$$
Zatem prędkość graniczna będzie równa:
$$
v_{gr} = \sqrt{\frac{mg}{k}}
$$

Używając wyprowadzonego powyżej wzoru, policzmy przykładową prędkość graniczną - dla skoczka o wadze 97 kg, przy założeniu, że współczynnik $k$ jest stały i równy $0.5 \cdot 1.22 \cdot 1.95 \cdot 0.5$, a przyspieszenie grawitacyjne $g$ wynosi 9,8 $\frac{m}{s^2}$. Dane dla $m$ i $k$, które wykorzystamy, pochodzą z artykułu "Differential Equations in the New Millenium: the Parachute Problem", którego autorami są Douglas B. Meade oraz Allan A. Struthers [2].

$$
v_{gr} = \sqrt{\frac{mg}{k}} = \sqrt{\frac{97 \cdot 9.8}{0.5 \cdot 1.22 \cdot 1.95 \cdot 0.5}} \approx 39.98 \quad \left[ \frac{m}{s^2} \right]
$$

Przyjrzyjmy się teraz, w jaki sposób będzie wyglądał wykres prędkości od czasu, gdzie prędkości będą wyznaczone innym, numerycznym sposobem -  metodą Eulera.
Przyjmujemy warunek początkowy $v(0) = 0$, masę skoczka wraz ze sprzętem 97 kg, krok czasowy 0.001 s i czas końcowy 20 s.

```python
import matplotlib.pyplot as plt

def eulers_method(m, v_0, dt, t_k):
    time_array = [0]
    velocity_array = [v_0]

    g = 9.8
    k = 0.5* 1.22* 1.95*0.5
    t = 0
    v = v_0
    while t <= t_k:
        v = v + (g - (k/m) * v**2) * dt
        velocity_array.append(v)
        t += dt
        time_array.append(t)

    return velocity_array, time_array

velocities, times = eulers_method(97, 0, 0.001, 20)
print("Prędkość końcowa: " + str(velocities[-1]))
plt.plot(times, velocities)
plt.xlabel("Czas (s)")
plt.ylabel("Prędkość (m/s)")
plt.title("Prędkość spadochroniarza w czasie - metoda Eulera")
plt.show()
```

Możemy zaobserwować, że wartość prędkości rośnie do około 39.97 m/s, po czym pozostaje stała. Jest to wynik zbliżony do wyniku otrzymanego za pomocą analitycznego sposobu obliczania wartości prędkości granicznej skoczka.

Zmierzymy czas wykonania obydwu funkcji: obliczającej wyniki metodą Eulera oraz nowej, obliczającej wyniki sposobem analitycznym.

```python
from math import sqrt, exp
import matplotlib.pyplot as plt

def analytic_solution_velocities_array(m, dt, t_k):
    t = 0
    k = 1.95*0.5
    g = 9.8
    velocities_array = []

    while t <= t_k+dt:
        velocity = sqrt((m*g)/k)*((exp(2*t*sqrt((k*g)/m))-1)/(exp(2*t*sqrt((k*g)/m))+1))
        velocities_array.append(velocity)
        t += dt

    return velocities_array
    
```

```python
from time import perf_counter

def time_measure(m, v_0, dt, tk):

    start_time_analytic = perf_counter()
    analytic_solution_velocities_array(m, dt, t_k)
    end_time_analytic = perf_counter()
    result_analytic = end_time_analytic - start_time_analytic

    start_time_euler = perf_counter()
    eulers_method(m, v_0, dt, t_k)
    end_time_euler = perf_counter()
    result_euler = end_time_euler - start_time_euler

    return result_analytic, result_euler

m = 97
v_0 = 0
t_k = 20
dt = [0.0001, 0.001, 0.01, 0.1]

for step in dt:
    result_analytic, result_euler = time_measure(m, v_0, step, t_k)
    print("Czas obliczenia wyników sposobem analitycznym dla kroku czasowego " + str(step) + ": " + str(result_analytic) + " s, czas obliczenia wyników metodą Eulera: " + str(result_euler) + " s.")
```

Analizując powyższe wyniki, możemy stwierdzić, że dla każdego kroku czasowego obliczenie wszystkich wyników metodą Eulera okazało się szybszym sposobem niż obliczanie wyników analitycznie. Przyczyną może być fakt, że każdy poprzednio obliczony wynik jest przechowywany w funkcji eulers_method jako $v$, a następnie wykonywane są z jego wykorzystaniem kolejne obliczenia, natomiast w rozwiązaniu analitycznym wszystkie wyniki muszą być obliczane osobno.

Powyższy model ze stałym współczynnikiem $k$ jest wystarczająco dobry, aby zaprezentować zagadnienie prędkości granicznej, jednak żeby model skoku spadochronowego był jeszcze bardziej zbliżony do rzeczywistości, należy rozpatrzyć także inne czynniki wpływające na dynamikę ruchu skoczka. Współczynnik $k$ w rzeczywistości nie jest stały, lecz zmienia się w czasie, a wpływa na niego między innymi gęstość powietrza, która z kolei jest zależna m. in. od ciśnienia atmosferycznego i temperatury - im dalej od powierzchni ziemi, tym mniejsza jest gęstość powietrza [1, 3, 6]. Poniższa, zmodyfikowana funkcja z metodą Eulera uwzględnia zmieniającą się gęstość powietrza i tym samym - zmienny współczynnik $k$.

Jako początkową wysokość, na której znajduje się skoczek, przyjmujemy 1200 m. Według informacji zawartych w artykule [2], jest to wysokość, na której trenuje skoki spadochronowe drużyna spadochroniarzy Air Force Academy.

```python
import matplotlib.pyplot as plt
import numpy as np

def atmospheric_density(height):
    # Model ISA (International Standard Atmosphere) 
    T0 = 288.15  # Temperatura na poziomie morza [K]
    L = 0.0065   # Lapse rate [K/m]
    P0 = 101325  # Ciśnienie na poziomie morza [Pa]
    R = 287.05   # Stała gazowa powietrza [J/(kg·K)]

    # Obliczanie temperatury na danej wysokości
    temperature = T0 - L * height

    # Obliczanie ciśnienia na danej wysokości
    pressure = P0 * (1 - L * height / T0)**(9.8 / (R * L))

    # Obliczanie gęstości powietrza na danej wysokości
    density = pressure / (R * temperature)

    return density

#metoda Eulera z uwzględnieniem zmieniającej się gęstości powietrza
def eulers_method_for_equation_with_changing_density(m, drag_coefficient, h_0, v_0, dt, t_k):
    time_array = [0]
    velocity_array = [v_0]
    heights_array = [h_0]
    density_array = [atmospheric_density(h_0)]

    t = 0
    v = v_0
    h = h_0
    g = 9.81

    while t <= t_k:
        density = atmospheric_density(h)
        drag_force = 0.5 * density * drag_coefficient * v**2
        v += (g - (drag_force/m)) * dt
        h -= v*dt
        t += dt

        time_array.append(t)
        velocity_array.append(v)
        heights_array.append(h)
        density_array.append(density)

    return velocity_array, heights_array, time_array, density_array 

velocities, heights, times, densities= eulers_method_for_equation_with_changing_density(97, 1.95*0.5, 1200, 0, 0.001, 70) 

plt.plot(times, velocities)
plt.xlabel("Czas (s)")
plt.xlim(0, 20)
plt.ylabel("Prędkość spadochroniarza (m/s)")
plt.title("Prędkość spadochroniarza w czasie - z uwzględnieniem zmiennej gęstości powietrza")
plt.show()
```

Na powyższym wykresie można zaobserwować, że skoczek osiąga prędkość graniczną, po czym wartość prędkości zaczyna maleć. Dlaczego tak się dzieje? Gęstość powietrza jest tym większa, im bliżej powierzchni się znajdujemy. Skoro gęstość powietrza się zwiększa, to rośnie także wartość siły oporu, a zatem maleje prędkość skoczka.

Przyjrzyjmy się również, jak zmienia się wysokość skoczka nad powierzchnią Ziemi w czasie.

```python
plt.plot(times, heights)
plt.xlabel("Czas (s)")
plt.ylabel("Wysokość, na której znajduje się spadochroniarz (m)")
plt.xlim(0, 40)
plt.ylim(0, 1200)
plt.title("Wysokość spadochroniarza w czasie - z uwzględnieniem zmiennej gęstości powietrza")
plt.show()
```

Na początku prędkość skoczka jest niewielka, dlatego zmiany wysokości są mniejsze niż w późniejszych sekundach skoku. Skoczek jednak w szybkim tempie zwiększa swoją prędkość i dalej zależność wysokości od czasu przypomina funkcję liniową.

W ostatniej części weźmiemy pod uwagę nie tylko zmieniającą się gęstość powietrza. Najważniejsze w skoku spadochronowym jest przecież otwarcie spadochronu! Dotychczas przyjmowaliśmy, że we współczynniku $k$ zmienia się tylko gęstość powietrza. Składa się na niego jednak także pole powierzchni ciała oraz pewien współczynnik, zależny np. od kształtu ciała. Inne pole powierzchni będzie miał skoczek wraz z ekwipunkiem, a inne - skoczek z otwartym spadochronem [2]. Przyjrzymy się, co się dzieje, kiedy do pewnego czasu $t_{open}$ skoczek spada bez otwartego spadochronu, następnie otwiera go w czasie $t_{open}$, po czym aż do momentu zetknięcia się z ziemią leci z otwartym spadochronem. Również w tym przypadku wykorzystane przez nas dane pochodzą z artykułu "Differential Equations in the New Millenium: the Parachute Problem" [2].

```python
import matplotlib.pyplot as plt
import numpy as np

def atmospheric_density(height):
    # Model ISA (International Standard Atmosphere) 
    T0 = 288.15  # Temperatura na poziomie morza [K]
    L = 0.0065   # Lapse rate [K/m]
    P0 = 101325  # Ciśnienie na poziomie morza [Pa]
    R = 287.05   # Stała gazowa powietrza [J/(kg·K)]

    # Obliczanie temperatury na danej wysokości
    temperature = T0 - L * height

    # Obliczanie ciśnienia na danej wysokości
    pressure = P0 * (1 - L * height / T0)**(9.8 / (R * L))

    # Obliczanie gęstości powietrza na danej wysokości
    density = pressure / (R * temperature)

    return density

#metoda Eulera z uwzględnieniem zmieniającej się gęstości powietrza
def eulers_method_for_equation_with_changing_density(m, drag_coefficient_1, drag_coefficient_2, h_0, v_0, dt, t_open, t_k):
    time_array = [0]
    velocity_array = [v_0]
    heights_array = [h_0]
    ###
    density_array = [atmospheric_density(h_0)]
    t = 0
    v = v_0
    h = h_0
    g = 9.81

    while t <= t_open:
        density = atmospheric_density(h)
        drag_force = 0.5 * density * drag_coefficient_1 * v**2
        v += (g - (drag_force/m)) * dt
        h -= v*dt
        t += dt

        time_array.append(t)
        velocity_array.append(v)
        heights_array.append(h)
        density_array.append(density)

    while t_open < t <= t_k:
        density = atmospheric_density(h)
        drag_force = 0.5 * density * drag_coefficient_2 * v**2
        v += (g - (drag_force/m)) * dt
        h -= v*dt
        t += dt

        time_array.append(t)
        velocity_array.append(v)
        heights_array.append(h)
        density_array.append(density)


    return velocity_array, heights_array, time_array, density_array 

drag_coefficient_2 = 43.8*1.33 + 0.35*1.78*0.1
velocities, heights, times, densities= eulers_method_for_equation_with_changing_density(97, 1.95*0.5, drag_coefficient_2, 1200, 0, 0.001, 13, 170)
print("Prędkość końcowa: " + str(velocities[-1]))
plt.plot(times, velocities)
plt.xlabel("Czas (s)")
plt.xlim(0, 20)
plt.ylabel("Prędkość spadochroniarza (m/s)")
plt.title("Prędkość spadochroniarza w czasie - z uwzględnieniem otwarcia spadochronu")
plt.show()
```

Od momentu rozpoczęcia skoku wartość prędkości skoczka rośnie aż do osiągnięcia prędkości granicznej. Po 13 sekundach lotu skoczek otwiera spadochron - wartość siły oporu natychmiastowo wzrasta, co znajduje również odzwierciedlenie na wykresie prędkości. Wartość prędkości maleje do 5.14 $\frac{m}{s}$. Jest to zgodna z rzeczywistością wartość - według artykułu [2], prędkość skoczka w momencie zetknięcia się z powierzchnią Ziemi wynosi od około 5.3 do 5.72 $\frac{m}{s}$, a według artykułu [5] - około 5.2 $\frac{m}{s}$.

```python
plt.plot(times, heights)
plt.xlabel("Czas (s)")
plt.ylabel("Wysokość, na której znajduje się spadochroniarz (m)")
plt.ylim(0, 1250)
plt.title("Wysokość spadochroniarza w czasie - z uwzględnieniem otwarcia spadochronu")
plt.show()
```

Na powyższym wykresie widzimy, że od momentu otwarcia spadochronu wysokość, na której znajduje się skoczek, zmienia się wolniej - spadochron spełnia swoje zadanie. Wartość siły oporu powietrza znacznie się zwiększyła, a wartość prędkości - zmalała. Dotarcie do ziemi zajmuje skoczkowi około 160 sekund - otrzymany przez nas wynik ponownie pokrywa się z danymi z artykułu [2].

Na koniec rozważmy jeszcze jeden przypadek - kiedy spadochron jest zawsze otwierany na wysokości 1000 m nad powierzchnią Ziemi. Skoki spadochronowe często odbywają się na wyższych wysokościach niż 1200 m, sięgających nawet 3000 m [4, 8, 9] - taką właśnie tym razem przyjmiemy wysokość początkową. 1000 m jest wysokością, na której można otworzyć spadochron i bezpiecznie wylądować [7].

```python
import matplotlib.pyplot as plt
import numpy as np

def atmospheric_density(height):
    # Model ISA (International Standard Atmosphere) 
    T0 = 288.15  # Temperatura na poziomie morza [K]
    L = 0.0065   # Lapse rate [K/m]
    P0 = 101325  # Ciśnienie na poziomie morza [Pa]
    R = 287.05   # Stała gazowa powietrza [J/(kg·K)]

    # Obliczanie temperatury na danej wysokości
    temperature = T0 - L * height

    # Obliczanie ciśnienia na danej wysokości
    pressure = P0 * (1 - L * height / T0)**(9.8 / (R * L))

    # Obliczanie gęstości powietrza na danej wysokości
    density = pressure / (R * temperature)

    return density

#metoda Eulera z uwzględnieniem zmieniającej się gęstości powietrza
def eulers_method_for_equation_with_changing_density(m, drag_coefficient_1, drag_coefficient_2, h_0, v_0, dt):
    time_array = [0]
    velocity_array = [v_0]
    heights_array = [h_0]
    density_array = [atmospheric_density(h_0)]
    t = 0
    v = v_0
    h = h_0
    g = 9.81

    while h>= 1000:
        density = atmospheric_density(h)
        drag_force = 0.5 * density * drag_coefficient_1 * v**2
        v += (g - (drag_force/m)) * dt
        h -= v*dt
        t += dt

        time_array.append(t)
        velocity_array.append(v)
        heights_array.append(h)
        density_array.append(density)

    while h<= 1000 and h>=0:
        density = atmospheric_density(h)
        drag_force = 0.5 * density * drag_coefficient_2 * v**2
        v += (g - (drag_force/m)) * dt
        h -= v*dt
        t += dt

        time_array.append(t)
        velocity_array.append(v)
        heights_array.append(h)
        density_array.append(density)


    return velocity_array, heights_array, time_array, density_array

drag_coefficient_2 = 43.8*1.33 + 0.35*1.78*0.1
velocities, heights, times, densities= eulers_method_for_equation_with_changing_density(97, 1.95*0.5, drag_coefficient_2, 3000, 0, 0.001)
print("Prędkość końcowa: " + str(velocities[-1]))
plt.plot(times, velocities)
plt.xlabel("Czas (s)")
plt.xlim(0, 100)
plt.ylabel("Prędkość spadochroniarza (m/s)")
plt.title("Prędkość spadochroniarza w czasie - z uwzględnieniem otwarcia spadochronu")
plt.show()

plt.plot(times, heights)
plt.xlabel("Czas (s)")
plt.ylabel("Wysokość, na której znajduje się spadochroniarz (m)")
plt.title("Wysokość spadochroniarza w czasie - z uwzględnieniem otwarcia spadochronu")
plt.show()
```

Z wykresu wysokości od czasu możemy odczytać, że tym razem skok spadochroniarza trwa ponad 200 sekund. Jego prędkość tuż przed zetknięciem się z ziemią wynosi $5.16\frac{m}{s}$.
